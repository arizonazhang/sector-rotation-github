```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from utils import load_data, cal_perf
import matplotlib.ticker as ticker
import matplotlib.dates as mdates
```

```python
sector_index = 'hsci'
code = 'hk'
window = 60
df_sectors, p, T = load_data(sector_index, code)
gamma_op = 2
gamma_reg = 0.0001
```

```python
country = "Hong Kong"
# time series as x-axis, tough issue.. matplotlib go to hell...
def allocation_plot(sector_index, sectornames, country, code, lag):
    i = 0
    df_sectors, p, T = load_data(sector_index, code)
    returns = df_sectors.iloc[window:, :p]
    fig, axes = plt.subplots(figsize= (20,12), ncols = 1, nrows = 2)
    plt.style.use('seaborn')
    for gamma_op in [2,4]:
        for gamma_reg in [0.00001]:
            weights = pd.read_csv(r".\sector-rotation-data\weights-quadprog_{}_{:.0f}_{:.0f}.csv".format(sector_index, gamma_op, -np.log10(gamma_reg)),index_col=0)
    #         weights.index = pd.to_datetime(weights.index, format="%Y-%m")
            ret_ls = [returns.iloc[i, :p].dot(weights.iloc[i]) for i in range(weights.shape[0])]
            ret_ls = np.array(ret_ls)
            avg_return = ret_ls.mean()
            annual_ex_return = (ret_ls + 1).prod() ** (12 / len(ret_ls)) - 1
            annual_vol = ret_ls.std() * np.sqrt(12)
            annual_sharpe = annual_ex_return / annual_vol
            s1 = "$\gamma = {} $; regularization: 0.00001".format(gamma_op)
            s2 = "(annual excess return: {:.2%}, annual volatility: {:.2%}, annual sharpe ratio: {:.4f})".format(
                annual_ex_return, annual_vol, annual_sharpe)
    #         weights.index = mdates.date2num(weights.index)
    #         axes[i].bar(x=weights.index,height=weights.iloc[:,0],color="r")
            weights.plot(kind='bar', stacked=True, title=s1, colormap='Paired', ax = axes[i])
    #         axes[i].xaxis.set_major_locator(mdates.MonthLocator(bymonth = 12))
    #         axes[i].xaxis.set_major_formatter(mdates.AutoDateLocator())
            axes[i].xaxis.set_major_locator(ticker.FixedLocator(locs=np.arange(8,len(weights.index),12)))
            axes[i].legend(sectornames, ncol=1, loc="center left", bbox_to_anchor = (1, 0.5), fontsize = 11, shadow = False, fancybox = False)
            axes[i].set_xlabel("")
            i += 1
    fig.suptitle("Sector Allocation ({})".format(country), fontsize = 16)        
    fig.tight_layout(rect = [0, 0.03, 1, 0.98])
    plt.savefig("allocation_{}.png".format(code))
    plt.show()
```

```python
sectornames = ["Energy", "Materials", "Industrials", "Consumer Discretionary", "Consumer Staples", "Health Care", "Financials",
               "Information technology", "Telecom", "Utilities"]
allocation_plot("csi", sectornames, "Shanghai-Shenzhen", "cn", lag = 10)
```

```python
sectornames = ["Basic Materials", "Consumer Goods", "COnsumer Services", "Financials", "Health Care", "Industrials", "Oil & Gas", 
              "Technology", "Telecom", "Utilities"]
allocation_plot("dj", sectornames, "U.S.", "us")
```
