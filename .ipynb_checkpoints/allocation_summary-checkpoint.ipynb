```python
# time series as x-axis, tough issue.. matplotlib go to hell...
def cal_jensen(X, Y):
    """
    calculate Jensen's alpha for a return series. 
    Y denotes the return series and X is the excess market return
    returns Jensen's alpha and its p-value
    """
    import statsmodels.api as sm
    X = pd.DataFrame(X)
    X['const']=1
    res = sm.OLS(Y, X).fit()
    alpha_p = res.pvalues['const']
    alpha = res.params['const']
    return alpha, alpha_p
```

```python
def cal_perf(code, method, useweekly, show_table = False):
    """
    calculate performance statistics of a method used for a certain market (code)
    code is the country code
    when show_table = True, the function prints a HTML table to the console
    """
    ret_ls = pd.read_csv(r".\performances\{}\{}_{}.csv".format("weekly" if useweekly else "monthly", code, method), index_col = 0)
    mkt_ls = pd.read_csv(r".\input\factor\ten_factor_vw_{}{}_5.csv".format(code, "_week" if useweekly else ""), index_col = 0, usecols = ["Week" if useweekly else "Month", "market", "rf"])
    ret_ls = ret_ls.join(mkt_ls)
    name = ret_ls.columns[1]
    values = ret_ls.iloc[:,1].unique()
    b_weight = ret_ls.iloc[:,2].unique()
    periods = 52 if useweekly else 12
    df = {}
    import scipy.stats
    for j, b in enumerate(b_weight):
        for i, pv in enumerate(values):
            _ret_ls = ret_ls.loc[(ret_ls[name] == pv)&(ret_ls.b_weight == b)]
            Y = _ret_ls.iloc[:,0]
            X = _ret_ls["market"] - _ret_ls["rf"]
            avg_return = Y.mean()
            cum_return = (Y + 1).prod() - 1
            annual_ex_return = (Y + 1).prod() ** (periods / len(Y)) - 1 if  cum_return > 0 else None
            annual_vol = Y.std() * np.sqrt(periods)
            annual_sharpe = annual_ex_return / annual_vol if cum_return > 0 else None
            alpha, alpha_p = cal_jensen(X, Y)
            df[('%.2f' %b, "%.4f" %pv)] = [cum_return, annual_ex_return, annual_vol, annual_sharpe, alpha, alpha_p]
    df = pd.DataFrame.from_dict(df,orient="index",columns=["Cum.Ret", "Ann.Ret", "Ann.Vol", 
                                                           "Ann.Sharpe", "Jensen's Alpha", "p-value"])
    df.index = pd.MultiIndex.from_tuples(list(df.index), names=["b", name])
    if show_table:
        display(df.style.format('{:.2%}').set_caption("Portfolio Performance ({})".format(code)))
    return df
```

```python
# date formatting only works after switching to plt methods..
def plot_one_weight(data, sectornames, periods, stats, title, ax, lag):
    """
    return a subplot of a time series of allocation data
    stats is the string containing performance statistics
    title is the subplot title
    """

    import matplotlib
    cmap = matplotlib.cm.get_cmap('Paired')

#     data.plot(kind='bar', stacked=True, colormap='Paired', ax=ax)
    for i in range(len(data.columns)):
        ax.bar(data.index, data.iloc[:,i], bottom = data.iloc[:,:i].sum(axis=1), width = 4, color = cmap((i+0.5)/len(data.columns)))

    ax.legend(sectornames, ncol=1, loc="center left", bbox_to_anchor=(1, 0.5), fontsize=11)
    ax.set_xlabel("")
    ax.set_xlim([data.index[0],data.index[-1]])
    ax.set_ylim([0, 1])
#     ax.set_xlim([data.index[0], data.index[-1]])
#     data.index = mdates.date2num(data.index)
#     ax.xaxis.set_major_locator(ticker.FixedLocator(locs=np.arange(lag, len(data.index), 12)))
    ax.xaxis.set_major_locator(mdates.MonthLocator(bymonth=(6,12)))
    ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
    ax.tick_params(rotation=90)
    
    ax.text(ax.get_xlim()[0]+0.05, ax.get_ylim()[1]+0.01, title,  ha="left", va = "bottom", fontweight = 'bold', fontsize = 12)
    ax.text(ax.get_xlim()[1]-0.05, ax.get_ylim()[1]+0.01, stats, ha = "right",va = "bottom", color = "gray", fontsize = 12)
    
    return ax
```

```python
def allocation_plot(sector_index, sectornames, country, 
                    code, method, name, values, 
                    show_table = False, lag = 9, useweekly = False,
                    save_fig = False, window=60, b_weight=[0.5, 2/3, 1]):
    """
    plots all allocation specified for a method with varying weight and gamma
    """
#     start = time.process_time()
    p = len(sectornames)
    perf = cal_perf(code, method, useweekly, show_table)
    ##print("cal_perf: {}".format(time.process_time()-start))
    
    # create subplots
    plt.style.use('seaborn')
    col = min(len(b_weight), len(values)) if len(b_weight) or len(values) else len(b_weight)
    row = max(len(b_weight), len(values)) if len(b_weight) or len(values) else len(b_weight)
    fig, axes = plt.subplots(figsize=(15*len(b_weight), 5*len(values)), ncols=col, nrows=row)
    
    # plot each subplots
    weights = pd.read_csv(r".\weights\{}\{}_{}.csv".format("weekly" if useweekly else "monthly", code,method),index_col = 0)
    weights.index = pd.to_datetime(weights.index)
    for i, pv in enumerate(values):
        for j, b in enumerate(b_weight):
            s1 = "$\gamma$ = {}, w = {:.0%}".format(str(pv), b)
            _perf = perf.loc[('%.2f' %b, "%.4f" %pv)].to_list()
            s2 = "\n Ann. Ex Return: {1:.2%}, Ann. Sharpe: {3:.4f}, Jensen's alpha: {4:.2%}({5:.2f})".format(*_perf)
#             if useweekly:
#                 _weights = weights.loc[(weights[name] == pv)&(weights.b_weight == b)].iloc[::4,:p]
#             else:
#                 _weights = weights.loc[(weights[name] == pv)&(weights.b_weight == b)].iloc[:,:p]
            _weights = weights.loc[(weights[name] == pv)&(weights.b_weight == b)].iloc[:,:p]
            # only one subplot
            if len(values) == 1 and len(b_weight) == 1:
                axes = plot_one_weight(_weights, sectornames, 52 if useweekly else 12 , s2, s1, axes, lag)
            # only one column of subplots
            elif len(values) == 1:
                axes[j] = plot_one_weight(_weights, sectornames, 52 if useweekly else 12 , s2, s1, axes[j], lag)
            elif len(b_weight) == 1:
                axes[i] = plot_one_weight(_weights, sectornames, 52 if useweekly else 12 , s2, s1, axes[i], lag)
            # several column of subplots
            else:
                axes[i,j] = plot_one_weight(_weights, s2, s1, axes[i,j], lag)
#             print("subplot: {}".format(time.process_time()-start))
    
    # subplot title
    fig.suptitle("Sector Allocation ({})".format(country), fontsize=16)
    fig.tight_layout(rect=[0, 0.03, 1, 0.98])
#     print("tight layout: {}".format(time.process_time()-start))
    
    if save_fig:
        plt.savefig(".\plots\{}_allocation_{}{}.png".format(code, method,"_week" if useweekly else "_month" ))
    plt.show()
#     print("tight layout: {}".format(time.process_time()-start))
```

```python
def comparison_plot(sector_index, sectornames, country, code, 
                    methods, name, value, b_weight, show_table = False,
                    lag = 11, useweekly=False, save_fig = False, window=60):
    """
    plots all allocation specified for a method with varying weight and gamma
    """
    p = len(sectornames)
    
    # create subplots
    plt.style.use('seaborn')
    fig, axes = plt.subplots(figsize=(15, 5*len(methods)), ncols=1, nrows=len(methods))
    
    # plot each subplots
    for i, method in enumerate(methods):
        perf = cal_perf(code, method, useweekly, show_table = show_table)
        if not useweekly:
            weights = pd.read_csv(r".\weights\monthly\{}_{}.csv".format(code,method),index_col = 0)
        else:
            weights = pd.read_csv(r".\weights\weekly\{}_{}.csv".format(code,method),index_col = 0)
        if "factor" in method:
            method = "factor"
        elif "capm" in method:
            method = "capm"
        elif "hist" in method:
            method = "historical"
        s1 = "{}: $\gamma$ = {}, w = {:.0%}".format(method, str(value), b_weight)
        _perf = perf.loc[('%.2f' %b_weight, "%.4f" %value)].to_list()
        s2 = "\n Ann. Vol: {2:.2%}, Ann. Sharpe: {3:.2f}, Jensen's Alpha:{4:.2%}({5:.2f})".format(*_perf)
        if useweekly:
            _weights = weights.loc[(weights[name] == value)&(weights.b_weight == b_weight)].iloc[::4,:p]
        else:
            _weights = weights.loc[(weights[name] == value)&(weights.b_weight == b_weight)].iloc[:,:p]
        axes[i] = plot_one_weight(_weights, sectornames, 13 if useweekly else 12, s2, s1, axes[i], lag)
    
    # subplot title
    fig.suptitle("Sector Allocation Comparison({})".format(country), fontsize=16)
    fig.tight_layout(rect=[0, 0.03, 1, 0.98])
    
    if save_fig:
        path = ".\plots\{}_allocation_cmp_{}{}.png".format(code, method, "_week" if useweekly else "_month")
        plt.savefig(path)
    plt.show()
```

## Long-short

### Factor Model
- E, V estimation：

$ E = \beta  \bar{f} + \gamma_{reg} I_n$ 

$ V = \beta  \Sigma_f  \beta' + \Sigma_\varepsilon  + \gamma_{reg} I_{n\times n}$ 

$ \gamma_{reg} = 0.0001, 0.001$

- calculation:

$ w = V^{-1} E $

$ w_i = \frac{w_i}{\sum_{j=1}^{N}w_j}$

- weighting adjustment: 

$w = b \times w + (1-b) \times \frac{1}{N} \text{ where } b = 0.5, 2/3, 1$

```python
params = {'method':'formula_factore',
         'name': 'gamma_reg',
         'values': [0.005, 0.001],
         'b_weight': [1],
         'show_table': True}
allocation_plot(**code_list_week['hk'], **params)
allocation_plot(**code_list_week['cn'], **params)
# allocation_plot(**code_list['us'], **params)
```

### CAPM

- E, V estimation：(only ONE factor in f: market excess return)

$ E = \beta  \bar{f} + \gamma_{reg} I_n$

$ V = \beta  \Sigma_f  \beta' + \gamma_{reg} I_{n\times n}$

$ \gamma_{reg} = 0.0001, 0.001$

- calculation:

$ w = V^{-1} E $

$ w_i = w_i/\sum_{j=1}^{N}w_j$

- weighting adjustment: 

$w = b \times w + (1-b) \times \frac{1}{N} \text{ where } b = 0.5, 2/3$

```python
params = {'method':'formula_capme',
         'name': 'gamma_reg',
         'values': [0.005, 0.001],
         'b_weight': [1],
         'show_table': True}
allocation_plot(**code_list_week['hk'], **params)
allocation_plot(**code_list_week['cn'], **params)
# allocation_plot(**code_list['us'], **params)
```
