```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
import matplotlib.ticker as mticker
```

```python
data = pd.read_csv(r'.\input\swindex.csv')
```

```python
data.shape
```

```python
data.columns = ['code', 'name', 'date', 'open', 'high', 'low', 'close', 'volume', 'turnover', 'change']
data.head()
data = data.dropna()
```

```python
data.loc[data.name=="化工",'name']
```

```python
data1 = data.pivot(index="date", columns='name', values='close')
data1.head()
```

```python
data1.index = pd.to_datetime(data1.index, format="%d/%m/%Y")
data1 = data1.sort_index()
# data1.index = pd.to_datetime(data1.index, format="%d/%m/%Y %H:%M")
```

```python
data1.apply(lambda x: pd.Series([x.dropna().index.min(), x.dropna().index.max()])).T
```

```python
week_data = data1.groupby([data1.index.year, data1.index.isocalendar().week]).tail(1)
print(week_data.index.weekday)
week_data.index = list(map(lambda x: x + np.timedelta64(4-x.weekday(), 'D'), week_data.index))
week_return = week_data.pct_change(fill_method=None)
week_return
# how to save? change col name?
```
